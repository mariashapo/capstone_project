```python
import pandas as pd
import numpy as np

import re
```

***
Functions def

```python
def sel_col(df,reg_exp):
    '''
    This function return a filtered dataframe that contains given (reg_exp) in the column name
    Census_Tract is set as index
    '''
    df['Census_Tract']=df['Census_Tract'].astype('int32')
    temp_df=df.set_index('Census_Tract',drop=True)
    new_df=temp_df.loc[:,[bool(re.search(reg_exp,col)) for col in temp_df]]
    
    return new_df
```

```python
def preview (df):
    '''
    This formula return basic EDA analysis for the input dataframe in the form of a new dataframe
    '''
    print(f'The dataframe shape is {df.shape}')
    preview_df=pd.DataFrame({'Data Types':df.dtypes,'Total Null Values':df.isna().sum(), 'Null Values Percentage':df.isna().sum()*100/df.shape[0],'Sample Value Head':df.head(1).T.iloc[:,0], 'Sample Value Tail':df.tail(1).T.iloc[:,0],'Sample Value':df.sample().T.iloc[:,0]})
    preview_df.rename_axis('Column_Name',inplace=True)
    return preview_df
```
